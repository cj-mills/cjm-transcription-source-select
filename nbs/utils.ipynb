## Extension Sets

Supported audio and video file extensions for filtering. These use dotted format (e.g. `.mp3`) to match `Path.suffix.lower()` directly.

```python
#| export
AUDIO_EXTENSIONS = frozenset({".mp3", ".wav", ".flac", ".aac", ".ogg", ".m4a", ".wma", ".opus"})
VIDEO_EXTENSIONS = frozenset({".mp4", ".mkv", ".avi", ".mov", ".webm", ".wmv", ".flv"})
MEDIA_EXTENSIONS = AUDIO_EXTENSIONS | VIDEO_EXTENSIONS
```

## File Type Detection

```python
#| export
def detect_file_type(
    file_path: str  # Path to the file
) -> Optional[str]:  # "audio", "video", or None if not a media file
    """Detect whether a file is audio, video, or not a media file based on extension."""
    ext = Path(file_path).suffix.lower()
    if ext in AUDIO_EXTENSIONS:
        return "audio"
    elif ext in VIDEO_EXTENSIONS:
        return "video"
    return None
```

```python
#| export
def is_media_file(
    file_path: str  # Path to the file
) -> bool:  # True if the file has a supported media extension
    """Check if a file has a supported audio or video extension."""
    return Path(file_path).suffix.lower() in MEDIA_EXTENSIONS
```

## Duration Formatting

```python
#| export
def format_duration(
    seconds: Optional[float]  # Duration in seconds, or None
) -> str:  # Formatted duration string (e.g. "1:23:45" or "42:17")
    """Format a duration in seconds as H:MM:SS or MM:SS."""
    if seconds is None:
        return "--:--"
    total = int(seconds)
    hrs, remainder = divmod(total, 3600)
    mins, secs = divmod(remainder, 60)
    if hrs > 0:
        return f"{hrs}:{mins:02d}:{secs:02d}"
    return f"{mins}:{secs:02d}"
```

## Tests

```python
# File type detection
assert detect_file_type("/path/to/file.mp3") == "audio"
assert detect_file_type("/path/to/file.MP4") == "video"
assert detect_file_type("/path/to/file.txt") is None
assert detect_file_type("/path/to/file.wav") == "audio"
assert detect_file_type("/path/to/file.mkv") == "video"

# Media file check
assert is_media_file("/path/to/file.mp3") == True
assert is_media_file("/path/to/file.mp4") == True
assert is_media_file("/path/to/file.txt") == False

# Duration formatting
assert format_duration(None) == "--:--"
assert format_duration(0) == "0:00"
assert format_duration(65) == "1:05"
assert format_duration(3661) == "1:01:01"
assert format_duration(2537) == "42:17"

# File size formatting (reused from cjm-file-discovery)
assert format_file_size(0) == "0 B"
assert format_file_size(512) == "512 B"
assert format_file_size(1048576) == "1.0 MB"
assert format_file_size(1073741824) == "1.0 GB"

print("All tests passed!")
```
